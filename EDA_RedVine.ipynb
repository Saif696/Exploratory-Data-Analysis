# About the Dataset

The Red Wine Quality Dataset is a popular dataset used for regression and classification problems in machine learning. It contains chemical properties of red wine samples and their quality ratings, which are determined by professional wine tasters. The dataset provides valuable insights into how different chemical components affect wine quality.

**Source:**

The dataset is available from the UCI Machine Learning Repository.

**Dataset Details:**


* `Number of Samples`: 1,599 

* `Number of Features`: 12 (11 input variables + 1 target variable)

* ``Features``:

    1. ``fixed acidity``: Acids contribute to wine's taste and preservation.

   2. ``volatile acidity``: High levels can lead to an unpleasant vinegar taste.

   3. ``citric acid``: Adds freshness and flavor.

   4. ``residual sugar``: Remaining sugar after fermentation; contributes to sweetness.

   5. ``chlorides``: Salt content.

   6. ``free sulfur dioxide``: Prevents microbial growth and oxidation.

   7. ``total sulfur dioxide``: Total SO₂ content.

   8. ``density``: Wine's density compared to water.

   9. ``pH: Acidity level``; lower values mean higher acidity.

   10. ``sulphates``: Contributes to wine's taste and preservation.

   11. ``alcohol``: Alcohol content by volume.

   12. ``quality``: Quality score given by wine tasters (integer between 0 and 10, inclusive).

**Objective:**
The goal is to explore the dataset, analyze the relationships between the features and the quality scores, and derive meaningful insights that could aid predictive modeling for wine quality evaluation.

# Step#3:
*   **Univariate Analysis**

```python
# Plot Histogram for all Features

df.hist(bins=15, figsize=(15,10), color='steelblue', edgecolor = 'black')
plt.suptitle('Feature Distributions')
plt.show()
```

```python
# Box plot analysis for Outliers

plt.figure(figsize=(15,10))
for i, column in enumerate(df.columns[:-1], 1):
    plt.subplot(4,3,i)
    sns.boxplot(df[column])
    plt.title(f"Boxplot for {column}")
plt.tight_layout()
plt.show()
```

# Step#4:
*   **Bivariate Analysis**

```python
# Correlation among all features
df.corr()
```

```python
# Correlation heatmap

plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')  # annot=true for showing each value on map
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Pairplot for Reltionships

sns.pairplot(df, hue='quality', corner=True)
plt.show()
```

# Step#5:
*   **Feature Engineering**


```python
# Group Quality scores in categories

bins = [0,4,6,10]
labels = ['Low', 'Medium', 'High']
df['quality_category'] = pd.cut(df['quality'], bins=bins, labels=labels)

# Distribution of quality categories

sns.countplot(df['quality_category'], palette='viridis')
plt.title('Distribution of Quality Categories')
plt.show()
```

```python
# Box Plot for alcohol and quality category

sns.boxplot(x='quality_category', y='alcohol', data=df, palette='viridis')
plt.title('Alcohol Content vs Quality Category')
plt.show()
```

## Insights and Conclusions

1. **Correlation Analysis**:
   - Alcohol content and density are positively correlated with wine quality.
   - Volatile acidity shows a negative correlation with wine quality.

2. **Outliers**:
   - Features like `volatile acidity` and `residual sugar` exhibit outliers, which might require preprocessing.

3. **Feature Distributions**:
   - Features like alcohol are skewed, and normalization or scaling might be necessary for predictive modeling.

4. **Quality Categories**:
   - Wines with higher alcohol content tend to belong to higher quality categories.

These insights provide a foundation for predictive modeling or further analysis.
